# Markars

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([34,36,38,42,44,46])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'g')
plt.show()
```
